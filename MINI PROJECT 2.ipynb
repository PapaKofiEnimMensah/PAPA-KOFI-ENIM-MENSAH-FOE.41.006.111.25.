SMART DEVICE MANAGEMENT SYSTEM

```python
# OOP Mini Project: Smart Device Management System
# Author: Papa

# Parent Class: SmartDevice

class SmartDevice:
    def __init__(self, name, device_id):
        if not device_id:
            raise ValueError("Device ID cannot be empty")
        self.name = name
        self.__device_id = device_id
        self.__power_status = False  # default OFF

    @property
    def device_id(self):
        return self.__device_id

    @property
    def power_status(self):
        return self.__power_status

    def turn_on(self):
        self.__power_status = True
        print(f"{self.name} is now ON.")

    def turn_off(self):
        self.__power_status = False
        print(f"{self.name} is now OFF.")

    def display_info(self):
        status = "ON" if self.__power_status else "OFF"
        print(f"Device: {self.name}, ID: {self.__device_id}, Power: {status}")



# Child Class: TemperatureSensor

class TemperatureSensor(SmartDevice):
    def __init__(self, name, device_id, temperature=25):
        super().__init__(name, device_id)
        self.temperature = temperature

    def read_temperature(self):
        print(f"{self.name} reads {self.temperature}°C.")



# Child Class: SecurityCamera

class SecurityCamera(SmartDevice):
    def __init__(self, name, device_id):
        super().__init__(name, device_id)
        self.recording_status = False

    def start_recording(self):
        self.recording_status = True
        print(f"{self.name} started recording.")

    def stop_recording(self):
        self.recording_status = False
        print(f"{self.name} stopped recording.")



# Child Class: SmartLight

class SmartLight(SmartDevice):
    def __init__(self, name, device_id, brightness=50):
        super().__init__(name, device_id)
        if not (0 <= brightness <= 100):
            raise ValueError("Brightness must be between 0 and 100")
        self.brightness = brightness

    def increase_brightness(self):
        if self.brightness < 100:
            self.brightness += 10
            print(f"{self.name} brightness increased to {self.brightness}.")
        else:
            print("Brightness already at maximum.")

    def decrease_brightness(self):
        if self.brightness > 0:
            self.brightness -= 10
            print(f"{self.name} brightness decreased to {self.brightness}.")
        else:
            print("Brightness already at minimum.")


# Object Creation

temp_sensor = TemperatureSensor("Living Room Sensor", "TS001", 28)
smart_light = SmartLight("Bedroom Light", "SL001", 70)
camera = SecurityCamera("Front Door Camera", "SC001")


# Demonstration of Outputs

print("\n--- Device Information ---")
temp_sensor.display_info()
smart_light.display_info()
camera.display_info()

print("\n--- Turning Devices On/Off ---")
smart_light.turn_on()
smart_light.turn_off()

print("\n--- Reading Temperature ---")
temp_sensor.read_temperature()

print("\n--- Adjusting Brightness ---")
smart_light.increase_brightness()
smart_light.decrease_brightness()

print("\n--- Recording Camera ---")
camera.start_recording()
camera.stop_recording()



# Menu-Driven Interface

def menu():
    while True:
        print("\n--- Smart Device Management Menu ---")
        print("1. Display Device Information")
        print("2. Turn Device On")
        print("3. Turn Device Off")
        print("4. Read Temperature")
        print("5. Adjust Brightness")
        print("6. Start Recording")
        print("7. Exit")

        choice = input("Enter choice (1-7): ")

        if choice == "1":
            temp_sensor.display_info()
            smart_light.display_info()
            camera.display_info()
        elif choice == "2":
            smart_light.turn_on()
        elif choice == "3":
            smart_light.turn_off()
        elif choice == "4":
            temp_sensor.read_temperature()
        elif choice == "5":
            smart_light.increase_brightness()
            smart_light.decrease_brightness()
        elif choice == "6":
            camera.start_recording()
            camera.stop_recording()
        elif choice == "7":
            print("Exiting system...")
            break
        else:
            print("Invalid choice. Try again.")
    # menu()
```
